```python
print("---- FULL MENU ----")

# unique categories
categories = set()

for item in menu:
    categories.add(menu[item]["category"])

# category wise print
for cat in categories:
    print(f"\n===== {cat} =====")

    for item, details in menu.items():
        if details["category"] == cat:
            status = "Available" if details["available"] else "Unavailable"
            print(f"{item}  ₹{details['price']:.2f}  [{status}]")
```

```python
print("\n---- MENU STATS ----")

# total items
total_items = len(menu)
print("Total items:", total_items)

# available items
available_count = 0
for details in menu.values():
    if details["available"]:
        available_count += 1

print("Available items:", available_count)

# most expensive item
max_item = ""
max_price = 0

for item, details in menu.items():
    if details["price"] > max_price:
        max_price = details["price"]
        max_item = item

print("Most expensive item:", max_item, "₹", max_price)

# items under 150
print("\nItems under ₹150:")

for item, details in menu.items():
    if details["price"] < 150:
        print(item, "₹", details["price"])
```

```python
cart = []
```

```python
def add_item(item_name, qty):
    if item_name not in menu:
        print(item_name, "does not exist in menu")
        return

    if not menu[item_name]["available"]:
        print(item_name, "is unavailable")
        return

    for item in cart:
        if item["item"] == item_name:
            item["quantity"] += qty
            print(item_name, "quantity updated")
            return

    cart.append({
        "item": item_name,
        "quantity": qty,
        "price": menu[item_name]["price"]
    })
    print(item_name, "added to cart")


def remove_item(item_name):
    for item in cart:
        if item["item"] == item_name:
            cart.remove(item)
            print(item_name, "removed from cart")
            return
    print(item_name, "not found in cart")


def update_quantity(item_name, qty):
    for item in cart:
        if item["item"] == item_name:
            item["quantity"] = qty
            print(item_name, "quantity updated to", qty)
            return
    print(item_name, "not found in cart")


def show_cart():
    print("\nCurrent Cart:")
    for item in cart:
        print(item)
```

```python
add_item("Paneer Tikka", 2)
show_cart()

add_item("Gulab Jamun", 1)
show_cart()

add_item("Paneer Tikka", 1)
show_cart()

add_item("Mystery Burger", 1)
show_cart()

add_item("Chicken Wings", 1)
show_cart()

remove_item("Gulab Jamun")
show_cart()
```

```python
print("\n====== Order Summary ======")

subtotal = 0

for item in cart:
    item_total = item["quantity"] * item["price"]
    subtotal += item_total
    print(f"{item['item']} x{item['quantity']} = ₹{item_total}")

gst = subtotal * 0.05
total = subtotal + gst

print("--------------------------")
print("Subtotal:", subtotal)
print("GST (5%):", gst)
print("Total Payable:", total)
print("==========================")
```

```python
import copy

inventory_backup = copy.deepcopy(inventory)

print("Backup created")
```

```python
# original changes
inventory["Paneer Tikka"]["stock"] = 5

print("\nOriginal Inventory:", inventory["Paneer Tikka"])
print("Backup Inventory:", inventory_backup["Paneer Tikka"])

# restore done
inventory = copy.deepcopy(inventory_backup)
```

```python
print("\n---- Updating Inventory ----")

for item in cart:
    name = item["item"]
    qty = item["quantity"]

    stock = inventory[name]["stock"]

    if stock >= qty:
        inventory[name]["stock"] -= qty
    else:
        print(f"Only {stock} available for {name}")
        inventory[name]["stock"] = 0
```

```python
print("\n---- Reorder Alerts ----")

for item, details in inventory.items():
    if details["stock"] <= details["reorder_level"]:
        print(f"Reorder Alert: {item} - only {details['stock']} left (level: {details['reorder_level']})")
```

```python
print("\n---- Final Inventory ----")
print(inventory)

print("\n---- Backup Inventory ----")
print(inventory_backup)
```

```python
print("---- Revenue Per Day ----")

revenue_per_day = {}

for date, orders in sales_log.items():
    total = 0
    for order in orders:
        total += order["total"]

    revenue_per_day[date] = total
    print(date, ":", total)
```

```python
best_day = max(revenue_per_day, key=revenue_per_day.get)

print("\nBest Selling Day:", best_day, "with revenue", revenue_per_day[best_day])
```

```python
print("\n---- Most Ordered Item ----")

item_count = {}

for orders in sales_log.values():
    for order in orders:
        for item in order["items"]:
            if item not in item_count:
                item_count[item] = 0
            item_count[item] += 1

most_ordered = max(item_count, key=item_count.get)

print("Most ordered item:", most_ordered)
```
